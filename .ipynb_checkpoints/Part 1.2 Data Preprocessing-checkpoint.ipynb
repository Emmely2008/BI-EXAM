```python
import re

def get_price(price_str):
    price_regexp = r"(?P<price>\d+)"
    if 'Free admission' in price_str:
        price = 0
    elif 'ratis' in price_str:
        price = 0
    else :
        m = re.search(price_regexp, price_str)
    try:
        price = int(m.group('price'))
    except:
        price = None
    return price      
        
    
```
